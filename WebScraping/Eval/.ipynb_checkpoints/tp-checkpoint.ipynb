```python
import pandas as pd


# df = pd.read_csv('Fifa_world_cup_matches.csv', usecols=['team1','team2','number of goals team1', 'number of goals team2'])
# df1= df.groupby(by="team1").sum()
# df2=df.groupby(by="team2").sum()
# df3 = df1 +df2
# df3

data = pd.read_csv("Fifa_world_cup_matches.csv")
# Sum goals per team
avg1 = data.groupby('team1')['number of goals team1'].sum()
avg2 = data.groupby('team2')['number of goals team2'].sum()
avg = avg1 + avg2
#print(avg)
#print(len(avg))
print(avg.sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt

# Graphic of the sum of goals per team
dfcountries = data.loc[:,'team1']
countries = list(dfcountries)
countries = list(dict.fromkeys(countries))
countries = sorted(countries)
names=countries
Titres = []
for res in avg:
    Titres.append(res)
plt.pie(Titres, labels=names, labeldistance=1.25);
plt.legend(bbox_to_anchor=(2,1.1),title="Countries")
plt.show()
```
