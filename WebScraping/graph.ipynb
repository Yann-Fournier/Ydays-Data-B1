```python
import pandas as pd
# Avec matplotlib
import numpy as np
from matplotlib import pyplot as plt



# Avec seaborn
import seaborn as sns
#seaborn.path.append('/Users/yann/opt/anaconda3/lib/python3.9/site-packages/seaborn/')

voiture = pd.read_csv("voiture.csv")
```

```python
voiture5 = voiture[voiture["id"] < 6]
Nom5 = voiture5["Nom"]
Prix5 = voiture5["Prix"]

#print(voiture)
#print(type(Nom5))
#print(len(Prix5))
```

```python
# En Barplot (Nom/prix)

#x = pd.Series(Prix5, index=Nom5) (test)

# Make a random dataset:
# height = [3, 12, 5, 18, 45]
# bars = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(Nom5))

# Create bars
plt.bar(y_pos, Prix5)

# Create names on the x-axis
plt.xticks(y_pos, Nom5)

# Show graphic
plt.show()
```

```python
# En scatterplot (Puissance_din/puissance fiscale)
sns.regplot(x=voiture["Puissance_din"], y=voiture["Puissance fiscale"])
```

```python
# PS : Ca fonctionne pas trop comme je veux
# En barplots impbriques (kilomtrage en fonction de ca premier main)

sns.set_theme(style="whitegrid")

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=voiture5, kind="bar",
    x="Nom", y="Kilometrage", hue="Premire Main",
    errorbar="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Kilomtrage (km)")
g.legend.set_title("")
```

```python
# Exemple avec des pingouins

sns.set_theme(style="whitegrid")

penguins = sns.load_dataset("penguins")

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=penguins, kind="bar",
    x="species", y="body_mass_g", hue="sex",
    errorbar="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Body mass (g)")
g.legend.set_title("")
```
