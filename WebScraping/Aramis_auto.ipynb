```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd
import datetime
import sqlite3
```

```python
# Garder la page ouverte
chrome_options = Options() 
chrome_options.add_experimental_option("detach", True)

# On recupere le controle du driver
DRIVER_PATH = "./chromedriver"
driver = webdriver.Chrome(DRIVER_PATH, options=chrome_options)
driver.implicitly_wait(5)

# Connection à une base de données
#conn = sqlite3.connect('ma_base.db')
```

```python
page_url = "https://www.aramisauto.com/achat/recherche?page=0"
driver.get(page_url)
#time.sleep(2)
driver.find_element(By.XPATH, "/html/body/div[1]/div/div/div/div/div/div[2]/button[2]").click()
#time.sleep(2)
```

```python
# on récupère le nombre de voiture a scrapper
nbr_voitures = int(driver.find_element(By.XPATH, "/html/body/div[3]/div[2]/form/div[2]/div/h2/span[1]").text)

# on recupere le nombre de page
nbr_pages = int(nbr_voitures / 20)

# on recupere le nombre de voitures sur la derniere page
nbr_voitures_derniere_page = nbr_voitures - (nbr_pages*20)

print(nbr_voitures)
print(nbr_pages)
print(nbr_voitures_derniere_page)

#driver.close()
#driver.quit()
```

```python
#Creation des tableaux qui ou seront stockées les données de chaque voiture
Nom = []
Prix = []
Mise_en_circulation = []
Premiere_main = []
Kilometrage = []
Carburant = []
BDV = []
Puissance_fiscale = []
Puissance_din = []
```

```python
for i in range(nbr_pages):
#for i in range(1):
#for i in range(nbr_pages-2, nbr_pages):

    try:
        # Garder la page ouverte
        # chrome_options = Options() 
        # chrome_options.add_experimental_option("detach", True)

        # # On recupere le controle du driver
        # DRIVER_PATH = "./chromedriver"
        # driver = webdriver.Chrome(DRIVER_PATH, options=chrome_options)
        # driver.implicitly_wait(5)

        page_url = "https://www.aramisauto.com/achat/recherche?page=" + str(i)
        driver.get(page_url)
        #time.sleep(2)
        #driver.find_element(By.XPATH, "/html/body/div[1]/div/div/div/div/div/div[2]/button[2]").click()
        #time.sleep(2)

        if i == nbr_pages :
            elem = nbr_voitures_derniere_page +1
        else :
            elem = 21

        for y in range(1, elem):
            #print(y)

            try:
                if i == 4 :
                    pass
                else :
                    driver.find_element(By.XPATH, f"/html/body/div[3]/div[3]/div/div/div[{y}]/div/div[2]/div[1]/a").click()
                    #time.sleep(2)

                    try:
                        nom = driver.find_element(By.XPATH, "/html/body/div[2]/div/div/div[1]/div[1]/div[2]/div/div[2]/div[2]/div[2]/div/div/div/div/div/div[1]/div[2]/div/h1/ul/li[1]").text
                                                
                    except:
                        nom = None

                    # LE prix peut ce trouver à deux endroits différents (s'il y a une promo)
                    try:
                        prix = driver.find_element(By.XPATH, "/html/body/div[2]/div/div/div[1]/div[1]/div[2]/div/div[2]/div[2]/div[2]/div/div/div/div/div/div[3]/div/div/div[1]/div/div").text
                        prix = int(prix.replace(" ", "")[:-1])
                    except:
                        prix = None
                    
                    if prix == None :
                        try:
                            prix = driver.find_element(By.XPATH, "/html/body/div[2]/div/div/div[1]/div[1]/div[2]/div/div[2]/div[2]/div[2]/div/div/div/div/div/div[4]/div/div/div[1]/div/div[1]").text
                            prix = int(prix.replace(" ", "")[:-1])
                        except:
                            prix = None 

                    try:
                        mise_en_circulation = driver.find_element(By.XPATH, "/html/body/div[2]/div/div/div[1]/div[1]/div[2]/div/div[2]/div[2]/div[3]/section/div/div/div/div[1]/div[3]/div/div[2]/div[2]").text
                    except:
                        mise_en_circulation = None


                    # Split le type_voiture en premiere_main et kilométrage
                    try:
                        type_voiture = driver.find_element(By.XPATH, "/html/body/div[2]/div/div/div[1]/div[1]/div[2]/div/div[2]/div[2]/div[2]/div/div/div/div/div/div[1]/div[2]/div/h1/ul/li[3]/span[1]").text
                    except:
                        type_voiture = None
                    
                    if type_voiture == "Voiture 0 km":
                        premiere_main = 1
                        kilometrage = 0
                    elif type_voiture == None :
                        pass
                    else :
                        cpt = 0
                        str1 = ""
                        str2 = ""
                        for j in range(len(type_voiture)):
                            if cpt < 2 :
                                str1 += type_voiture[j]
                            elif cpt >= 2 :
                                str2 += type_voiture[j]
                            if type_voiture[j] == " ":
                                cpt += 1
                        premiere_main = 0
                        kilometrage = int(str2.replace(" ", "")[:-2])


                    try:
                        carburant = driver.find_element(By.XPATH, "/html/body/div[2]/div/div/div[1]/div[1]/div[2]/div/div[2]/div[2]/div[2]/div/div/div/div/div/div[1]/div[2]/div/div[3]/div/div/ul/li[2]/span[2]").text
                    except:
                        carburant = None

                    try:
                        bdv = driver.find_element(By.XPATH, "/html/body/div[2]/div/div/div[1]/div[1]/div[2]/div/div[2]/div[2]/div[2]/div/div/div/div/div/div[1]/div[2]/div/div[3]/div/div/ul/li[1]/span[2]").text
                        cpt2 = 0
                        str3 = ""
                        for u in range(len(bdv)):
                            if cpt2 < 2:
                                str3 += bdv[u]
                            elif bdv[u] == " ":
                                cpt2 += 1
                        bdv = str3     
                    except:
                        bdv = None

                    # La puissance fiscale peut ce trouver à deux endroit différent
                    try:
                        puissance_fiscale = driver.find_element(By.XPATH, "/html/body/div[2]/div/div/div[1]/div[1]/div[2]/div/div[2]/div[7]/div/section/div/div/div/div/div[5]/div/div/div/ul/li[3]/span").text
                        puissance_fiscale = int(puissance_fiscale.replace(" ", "")[1:][:-2])
                    except:
                        puissance_fiscale = None

                    if puissance_fiscale == None :
                        try:
                            puissance_fiscale = driver.find_element(By.XPATH, "/html/body/div[2]/div/div/div[1]/div[1]/div[2]/div/div[2]/div[7]/div/section/div/div/div/div/div[4]/div/div/div/ul/li[3]/span").text
                            puissance_fiscale = int(puissance_fiscale.replace(" ", "")[1:][:-2])
                        except:
                            puissance_fiscale = None

                    # La puissance din peut ce trouver a deux endroit différent
                    try:
                        puissance_din = driver.find_element(By.XPATH, "/html/body/div[2]/div/div/div[1]/div[1]/div[2]/div/div[2]/div[7]/div/section/div/div/div/div/div[5]/div/div/div/ul/li[2]/span").text
                        puissance_din = int(puissance_din.replace(" ", "")[1:][:-2])
                    except:
                        puissance_din = None
                    
                    if puissance_din == None :
                        try:
                            puissance_din = driver.find_element(By.XPATH,"/html/body/div[2]/div/div/div[1]/div[1]/div[2]/div/div[2]/div[7]/div/section/div/div/div/div/div[4]/div/div/div/ul/li[2]/span").text
                            puissance_din = int(puissance_din.replace(" ", "")[1:][:-2])
                        except:
                            puissance_din = None




                    Nom.append(nom)
                    Prix.append(prix)
                    Mise_en_circulation.append(mise_en_circulation)
                    Premiere_main.append(premiere_main)
                    Kilometrage.append(kilometrage)
                    Carburant.append(carburant)
                    BDV.append(bdv)
                    Puissance_fiscale.append(puissance_fiscale)
                    Puissance_din.append(puissance_din)

                    # print(nom)
                    # print(prix)
                    # print(mise_en_circulation)
                    # print(premiere_main)
                    # print(kilometrage)
                    # print(carburant)
                    # print(bdv)
                    # print(puissance_fiscale)
                    # print(puissance_din)
                   

                    driver.back()
                    driver.execute_script("window.history.go(-1)")
                    #time.sleep(2)

            except:
                pass
        
        #driver.close()
        #driver.quit()

    except:
        pass

driver.close()
driver.quit()
```

```python
df = pd.DataFrame({'Nom': Nom, 'Prix': Prix, 'Mise_en_circulation': Mise_en_circulation, 'Première Main': Premiere_main, 'Kilometrage': Kilometrage, 'Carburant': Carburant, 'Boîte de vitesse': BDV, 'Puissance fiscale' : Puissance_fiscale, 'Puissance_din': Puissance_din})
df.to_csv('Aramis_auto4.csv', index=False, encoding='utf-8')
#print(df)
```
